```python
from pandas import DataFrame
import matplotlib.pyplot as plt
%matplotlib inline
```

### 2.1 Model, trained on dataset with initial class balance (~1:5)

```python
# Test on the same domain (movie reviews, Amazon test)
data = DataFrame([['20k', 0.9145, 0.62],
                 ['100k', 0.9386,  0.75],
                 ['200k', 0.9444, 0.78],
                 ['500k', 0.9482, 0.80],
                 ['1kk', 0.9499, 0.81],
                 ['1.3kk', 0.9501, 0.81]
                 ], 
               columns = ['dataset', 'acc', 'f1'])

# plot it
plt.figure(figsize=((8,4.5)))
plt.title("Learning curve (initial balance): Amazon Movies Test")
xs = [i for i, _ in enumerate(data.acc)]
plt.plot(xs, data.acc, 'g-', label='Accuracy')
plt.plot(xs, data.f1, 'r-', label='F1 score (negative class)')
plt.xlabel("Train size")
plt.xticks(xs, data.dataset)
plt.legend()
plt.show()
```

### 3. Model, trained on varying balance dataset  (1:1, ..., 1:5)

```python
# Test on the same domain (movie reviews, Amazon test), 50k negative and 50k*x positive, x in [1,1.5,2.5,5]
data = DataFrame([['1:1.0',  0.9057, 0.73],
                 ['1:1.5', 0.9246,  0.76],
                 ['1:2.5', 0.9405, 0.80],
                 ['1:5.0', 0.9470, 0.80]], 
               columns = ['dataset', 'acc', 'f1'])

# plot it
plt.figure(figsize=((8,4.5)))
plt.title("Learning curve (varying balance): Amazon Movies Test")
xs = [i for i, _ in enumerate(data.acc)]
plt.plot(xs, data.acc, 'g-', label='Accuracy')
plt.plot(xs, data.f1, 'r-', label='F1 score (negative class)')
plt.xlabel("Train size, 1:k")
plt.xticks(xs, data.dataset)
plt.legend()
plt.show()
```
